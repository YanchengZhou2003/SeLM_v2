```python
# =========================
# Step 0: 依赖
# =========================
import os
import re
from typing import Dict, List, Any
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# =========================
# Step 1: 文件名生成层（新命名）
#   目标文件名：N{N}_Ndynbr{Ndynbr}_h{h}_tp{tp}.log
# =========================
def make_log_filename(N: int, Ndynbr: int, h: int, tp: int, prefix: str = "N") -> str:
    return f"{prefix}{N}_Ndynbr{Ndynbr}_h{h}_tp{tp}.log"

def make_log_filenames(N_list, Ndynbr_list, h_list, tp_list, prefix="N") -> List[str]:
    return [
        make_log_filename(N, Ndynbr, h, tp, prefix)
        for N in N_list for Ndynbr in Ndynbr_list for h in h_list for tp in tp_list
    ]

def list_existing_logs(log_dir: str, candidates: List[str]) -> List[str]:
    return [os.path.join(log_dir, f) for f in candidates if os.path.exists(os.path.join(log_dir, f))]

# =========================
# Step 2: 解析引擎层（单文件）
#   日志格式与你给的“新格式”一致：
#     epoch k summary: train_cos_loss : x, vocab_cro_loss : y,
#     train loss: a, accuracy: b
#     ...
#     epoch k summary: valid_cos_loss : v,
#     valid loss: c, accuracy: d
# =========================
FNAME_RE = re.compile(r"^N(?P<N>\d+)_Ndynbr(?P<Ndynbr>\d+)_h(?P<h>\d+)_tp(?P<tp>\d+)\.log$")

FLOAT = r"[-+]?(?:\d+(?:\.\d*)?|\.\d+)(?:[eE][-+]?\d+)?"
TRAIN_SUMMARY_RE = re.compile(r"^epoch\s+(\d+)\s+summary:\s+(.*)$")
TRAIN_KV_RE      = re.compile(rf"(\w+_loss)\s*:\s*({FLOAT})")
TRAIN_LINE_RE    = re.compile(r"^train loss:\s*(" + FLOAT + r")\s*,\s*accuracy:\s*(" + FLOAT + r")")
VALID_SUMMARY_RE = re.compile(r"^epoch\s+(\d+)\s+summary:\s+.*(valid_cos_loss)\s*:\s*(" + FLOAT + r")")
VALID_LINE_RE    = re.compile(r"^valid loss:\s*(" + FLOAT + r")\s*,\s*accuracy:\s*(" + FLOAT + r")")

def parse_filename_meta(filename: str) -> Dict[str, Any]:
    m = FNAME_RE.match(filename)
    if not m:
        return {"file": filename, "N": None, "Ndynbr": None, "h": None, "tp": None}
    return {
        "file": filename,
        "N": int(m.group("N")),
        "Ndynbr": int(m.group("Ndynbr")),
        "h": int(m.group("h")),
        "tp": int(m.group("tp")),
    }

def parse_log_file(filepath: str) -> Dict[str, Any]:
    meta = parse_filename_meta(os.path.basename(filepath))
    series: Dict[str, List[float]] = {
        "train_cos_loss": [], "vocab_cro_loss": [],
        "train_loss": [], "train_acc": [],
        "valid_cos_loss": [], "valid_loss": [], "valid_acc": []
    }
    epochs_train: List[int] = []
    epochs_valid: List[int] = []
    last_epoch_seen: int | None = None

    with open(filepath, "r") as f:
        for line in f:
            line = line.strip()

            mt = TRAIN_SUMMARY_RE.match(line)
            if mt:
                ep = int(mt.group(1))
                last_epoch_seen = ep
                kv_part = mt.group(2)
                for k, v in TRAIN_KV_RE.findall(kv_part):
                    if k in series:
                        series[k].append(float(v))
                epochs_train.append(ep)
                continue

            mt = TRAIN_LINE_RE.match(line)
            if mt:
                t_loss, t_acc = float(mt.group(1)), float(mt.group(2))
                series["train_loss"].append(t_loss)
                series["train_acc"].append(t_acc)
                continue

            mv = VALID_SUMMARY_RE.match(line)
            if mv:
                ep_v = int(mv.group(1))
                last_epoch_seen = ep_v
                val_cos = float(mv.group(3))
                series["valid_cos_loss"].append(val_cos)
                epochs_valid.append(ep_v)
                continue

            mv = VALID_LINE_RE.match(line)
            if mv:
                v_loss, v_acc = float(mv.group(1)), float(mv.group(2))
                series["valid_loss"].append(v_loss)
                series["valid_acc"].append(v_acc)
                if last_epoch_seen is not None and (not epochs_valid or epochs_valid[-1] != last_epoch_seen):
                    epochs_valid.append(last_epoch_seen)
                continue

    return {"meta": meta, "series": series, "epochs_train": epochs_train, "epochs_valid": epochs_valid}

# =========================
# Step 3: 批量收集层
# =========================
def parse_multiple_logs(filepaths: List[str]) -> List[Dict[str, Any]]:
    runs = []
    for p in filepaths:
        try:
            runs.append(parse_log_file(p))
            print(f"Parsed {p}")
        except Exception as e:
            runs.append({"meta": {"file": os.path.basename(p), "N": None, "Ndynbr": None, "h": None, "tp": None},
                         "series": {}, "error": str(e)})
    return runs

# =========================
# Step 4: 聚合与特征派生层
# =========================
def build_final_metrics_table(runs: List[Dict[str, Any]]) -> pd.DataFrame:
    rows: List[Dict[str, Any]] = []
    for r in runs:
        meta = r.get("meta", {})
        ser  = r.get("series", {})

        def last_or_nan(lst):
            return lst[-1] if lst else float("nan")

        rows.append({
            "file": meta.get("file"),
            "N": meta.get("N"),
            "Ndynbr": meta.get("Ndynbr"),
            "h": meta.get("h"),
            "tp": meta.get("tp"),
            "final_train_loss": last_or_nan(ser.get("train_loss", [])),
            "final_valid_loss": last_or_nan(ser.get("valid_loss", [])),
            "final_train_cos_loss": last_or_nan(ser.get("train_cos_loss", [])),
            "final_vocab_cro_loss": last_or_nan(ser.get("vocab_cro_loss", [])),
            "final_valid_cos_loss": last_or_nan(ser.get("valid_cos_loss", [])),
        })
    return pd.DataFrame(rows)

# =========================
# Step 5: 可视化层（按 Ndynbr 作曲线，横轴 N，min over (h,tp)）
#   2.1 以 training loss 取最小；2.2 以 validation loss 取最小
#   左图：gap；右图：train(实线)/valid(虚线)，颜色与左图一致
#   控制台打印每个 (N, Ndynbr) 选中的 (h,tp)
# =========================
def _select_min_over_htp(dfN: pd.DataFrame, Ndynbr: int, criterion: str):
    """在固定 (N, Ndynbr) 子表上，按 criterion ∈ {'final_train_loss','final_valid_loss'} 选出一行。"""
    sub = dfN[dfN["Ndynbr"] == Ndynbr]
    if sub.empty or sub[criterion].isna().all():
        return None
    idx = sub[criterion].idxmin()
    row = sub.loc[idx]
    return {
        "N": row["N"],
        "Ndynbr": row["Ndynbr"],
        "h": int(row["h"]) if pd.notna(row["h"]) else None,
        "tp": int(row["tp"]) if pd.notna(row["tp"]) else None,
        "final_train_loss": row["final_train_loss"],
        "final_valid_loss": row["final_valid_loss"],
        "gap": row["final_valid_loss"] - row["final_train_loss"],
    }

def plot_gap_and_losses_vs_N_groupby_Ndynbr(
    df: pd.DataFrame,
    criterion: str = "final_train_loss",
    title_suffix: str = "(min over h,tp by training loss)"
):
    """
    横轴 N；每条曲线代表一个 Ndynbr；在 (h,tp) 上按 criterion 取最小。
    左：gap；右：train/valid（实线/虚线，同色）。
    criterion ∈ {"final_train_loss", "final_valid_loss"}。
    返回用于复核的明细 DataFrame（包含选中的 h,tp）。
    """
    assert criterion in {"final_train_loss", "final_valid_loss"}

    Ndyn_list = sorted(df["Ndynbr"].dropna().unique())
    N_list = sorted(df["N"].dropna().unique())
    # 颜色从浅到深：Ndynbr 由小到大
    colors = plt.cm.Blues(np.linspace(0.35, 1.0, len(Ndyn_list)))
    cmap = {nd: colors[i] for i, nd in enumerate(Ndyn_list)}

    # 组装每条 Ndynbr 曲线的数据
    records_all = []  # 收集所有点，供打印和右图
    for N in N_list:
        dfN = df[df["N"] == N]
        for nd in Ndyn_list:
            pick = _select_min_over_htp(dfN, nd, criterion)
            if pick is not None:
                records_all.append(pick)

    if not records_all:
        print("No data to plot.")
        return pd.DataFrame()

    use_df = pd.DataFrame(records_all)
    # 控制台汇报 (N, Ndynbr) -> (h,tp)
    print("\n[Selected (h,tp) per (N, Ndynbr) by {}]".format(
        "train" if criterion == "final_train_loss" else "valid"))
    print(use_df.sort_values(["Ndynbr", "N"])[["N", "Ndynbr", "h", "tp",
                                               "final_train_loss", "final_valid_loss", "gap"]])

    # 绘图：左右子图
    fig, (axL, axR) = plt.subplots(1, 2, figsize=(13, 5), sharex=True)
    for nd in Ndyn_list:
        g = use_df[use_df["Ndynbr"] == nd].sort_values("N")
        if g.empty: 
            continue
        color = cmap[nd]
        # 左：gap
        axL.plot(g["N"], g["gap"], marker="o", color=color, label=f"Ndynbr={nd}")
        # 右：train/valid
        axR.plot(g["N"], g["final_train_loss"], marker="o", color=color, ls="-",  label=f"Ndynbr={nd} train")
        axR.plot(g["N"], g["final_valid_loss"], marker="o", color=color, ls="--", label=f"Ndynbr={nd} valid")

    axL.set_xlabel("N"); axL.set_ylabel("gap = valid - train")
    ttl = "Generalization gap vs N " + title_suffix
    axL.set_title(ttl); axL.grid(True, ls="--"); axL.legend()

    axR.set_xlabel("N"); axR.set_ylabel("loss")
    axR.set_title("Train/Valid loss vs N " + title_suffix)
    axR.grid(True, ls="--"); axR.legend(ncol=2, fontsize=8)

    plt.tight_layout()
    plt.show()
    return use_df

# 便捷包装：2.1 和 2.2
def plot_gap_min_by_training(df: pd.DataFrame):
    return plot_gap_and_losses_vs_N_groupby_Ndynbr(
        df, criterion="final_train_loss",
        title_suffix="(min over h,tp by training loss)"
    )

def plot_gap_min_by_validation(df: pd.DataFrame):
    return plot_gap_and_losses_vs_N_groupby_Ndynbr(
        df, criterion="final_valid_loss",
        title_suffix="(min over h,tp by validation loss)"
    )

# =========================
# Step 6: 示例（按需启用）
# =========================
if __name__ == "__main__":
    # 目录按你的启动脚本设置
    log_dir = "."

    N_list      = [1024, 2048, 4096, 8192, 16384]
    Ndynbr_list = [65, 130, 195, 260, 325]      # 升序用于“浅→深”配色
    h_list      = [14, 18, 22]
    tp_list     = [2, 4, 8]

    candidates = make_log_filenames(N_list, Ndynbr_list, h_list, tp_list)
    filepaths  = list_existing_logs(log_dir, candidates)
    runs       = parse_multiple_logs(filepaths)

    df_final = build_final_metrics_table(runs)
    display(df_final.sort_values(["N", "Ndynbr", "h", "tp"]))

    # 2.1 以 training loss 取 min
    df_sel_train = plot_gap_min_by_training(df_final)

    # 2.2 以 validation loss 取 min
    df_sel_valid = plot_gap_min_by_validation(df_final)
```
