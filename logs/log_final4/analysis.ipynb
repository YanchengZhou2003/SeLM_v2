```python
# =========================
# Step 0: 依赖
# =========================
import os
import re
from typing import Dict, List, Any
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 120

# =========================
# Step 1: 文件名生成 / 收集
# =========================
def make_log_filename(N: int, ratio_dyn: str | float, step_dyn: int, prefix: str = "N") -> str:
    return f"{prefix}{N}_rd{ratio_dyn}_sd{step_dyn}.log"

def make_log_filenames(N_list, ratio_list, step_list, prefix="N") -> List[str]:
    return [make_log_filename(N, r, s, prefix) for N in N_list for r in ratio_list for s in step_list]

def list_existing_logs(log_dir: str, candidates: List[str]) -> List[str]:
    return [os.path.join(log_dir, f) for f in candidates if os.path.exists(os.path.join(log_dir, f))]

# =========================
# Step 2: 解析引擎（单文件）
# =========================
FNAME_RE = re.compile(r"^N(?P<N>\d+)_rd(?P<rd>[0-9.]+)_sd(?P<sd>\d+)\.log$")

FLOAT = r"[-+]?(?:\d+(?:\.\d*)?|\.\d+)(?:[eE][-+]?\d+)?"

TRAIN_SUMMARY_RE = re.compile(
    r"^epoch\s+(\d+)\s+summary:\s+"
    r"dyn_dyn_cos_loss:\s*(" + FLOAT + r")\s*,\s*"
    r"dyn_sta_cos_loss:\s*(" + FLOAT + r")\s*,\s*"
    r"total_cos_loss\s*:\s*(" + FLOAT + r")\s*,\s*"
    r"vocab_cro_loss\s*:\s*(" + FLOAT + r")\s*,\s*$"
)
TRAIN_LINE_RE = re.compile(r"^train loss:\s*(" + FLOAT + r")\s*,\s*train acc:\s*(" + FLOAT + r")\s*$")
VALID_LINE_RE = re.compile(r"^valid loss:\s*(" + FLOAT + r")\s*,\s*valid acc:\s*(" + FLOAT + r")\s*$")

def parse_filename_meta(filename: str) -> Dict[str, Any]:
    m = FNAME_RE.match(filename)
    if not m:
        return {"file": filename, "N": None, "ratio_dyn": None, "step_dyn": None}
    return {
        "file": filename,
        "N": int(m.group("N")),
        "ratio_dyn": float(m.group("rd")),
        "step_dyn": int(m.group("sd")),
    }

def parse_log_file(filepath: str) -> Dict[str, Any]:
    meta = parse_filename_meta(os.path.basename(filepath))
    series: Dict[str, List[float]] = {
        "dyn_dyn_cos_loss": [], "dyn_sta_cos_loss": [], "total_cos_loss": [], "vocab_cro_loss": [],
        "train_loss": [], "train_acc": [], "valid_loss": [], "valid_acc": [],
    }
    with open(filepath, "r") as f:
        for line in f:
            line = line.strip()

            ms = TRAIN_SUMMARY_RE.match(line)
            if ms:
                series["dyn_dyn_cos_loss"].append(float(ms.group(2)))
                series["dyn_sta_cos_loss"].append(float(ms.group(3)))
                series["total_cos_loss"].append(float(ms.group(4)))
                series["vocab_cro_loss"].append(float(ms.group(5)))
                continue

            mt = TRAIN_LINE_RE.match(line)
            if mt:
                series["train_loss"].append(float(mt.group(1)))
                series["train_acc"].append(float(mt.group(2)))
                continue

            mv = VALID_LINE_RE.match(line)
            if mv:
                series["valid_loss"].append(float(mv.group(1)))
                series["valid_acc"].append(float(mv.group(2)))
                continue

    return {"meta": meta, "series": series}

# =========================
# Step 3: 批量解析
# =========================
def parse_multiple_logs(filepaths: List[str]) -> List[Dict[str, Any]]:
    runs = []
    for p in filepaths:
        try:
            runs.append(parse_log_file(p))
            print(f"Parsed {p}")
        except Exception as e:
            runs.append({"meta": {"file": os.path.basename(p), "N": None, "ratio_dyn": None, "step_dyn": None},
                         "series": {}, "error": str(e)})
    return runs

# =========================
# Step 4: 聚合（末尾指标）
# =========================
def build_final_table(runs: List[Dict[str, Any]]) -> pd.DataFrame:
    cols = ["file","N","ratio_dyn","step_dyn",
            "final_train_loss","final_valid_loss",
            "final_train_acc","final_valid_acc",
            "final_total_cos_loss","final_vocab_cro_loss",
            "final_dyn_dyn_cos_loss","final_dyn_sta_cos_loss"]
    if not runs:
        return pd.DataFrame(columns=cols)

    def last_or_nan(lst):
        return lst[-1] if lst else float("nan")

    rows = []
    for r in runs:
        meta = r.get("meta", {})
        s    = r.get("series", {})
        rows.append({
            "file": meta.get("file"),
            "N": meta.get("N"),
            "ratio_dyn": meta.get("ratio_dyn"),
            "step_dyn": meta.get("step_dyn"),
            "final_train_loss": last_or_nan(s.get("train_loss", [])),
            "final_valid_loss": last_or_nan(s.get("valid_loss", [])),
            "final_train_acc":  last_or_nan(s.get("train_acc", [])),
            "final_valid_acc":  last_or_nan(s.get("valid_acc", [])),
            "final_total_cos_loss": last_or_nan(s.get("total_cos_loss", [])),
            "final_vocab_cro_loss": last_or_nan(s.get("vocab_cro_loss", [])),
            "final_dyn_dyn_cos_loss": last_or_nan(s.get("dyn_dyn_cos_loss", [])),
            "final_dyn_sta_cos_loss": last_or_nan(s.get("dyn_sta_cos_loss", [])),
        })
    return pd.DataFrame(rows, columns=cols)

# =========================
# Step 5: 绘图（非交互，逐张画，纵轴一致）
# =========================
def _color_map_for_ratios(ratio_values):
    ratios_sorted = sorted(ratio_values)
    colors = plt.cm.Blues(np.linspace(0.3, 1.0, len(ratios_sorted)))
    return {r: colors[i] for i, r in enumerate(ratios_sorted)}

def plot_all_steps_static(df: pd.DataFrame):
    """
    逐个 step_dyn 出图；左：gap(train-valid)，右：train/valid。
    所有左图共享同一 y 轴范围；所有右图共享同一 y 轴范围。
    """
    if df.empty:
        print("No data to plot.")
        return

    df = df.copy()
    df["gap"] = df["final_valid_loss"] - df["final_train_loss"]

    # 颜色映射对所有图保持一致
    cmap = _color_map_for_ratios(df["ratio_dyn"].dropna().unique())

    # 全局 y 轴范围（保证纵轴一致性）
    gap_min = df["gap"].min()
    gap_max = df["gap"].max()
    loss_min = np.min([df["final_train_loss"].min(), df["final_valid_loss"].min()])
    loss_max = np.max([df["final_train_loss"].max(), df["final_valid_loss"].max()])

    gap_pad  = 0.05 * (gap_max - gap_min + 1e-12)
    loss_pad = 0.05 * (loss_max - loss_min + 1e-12)
    gap_ylim  = (gap_min - gap_pad,  gap_max + gap_pad)
    loss_ylim = (loss_min - loss_pad, loss_max + loss_pad)

    for step in sorted(df["step_dyn"].dropna().unique()):
        df_sel = df[df["step_dyn"] == step].sort_values(["ratio_dyn", "N"])
        if df_sel.empty:
            continue

        fig, (axL, axR) = plt.subplots(1, 2, figsize=(12, 5), sharex=False)

        # 左：gap
        for ratio, grp in df_sel.groupby("ratio_dyn"):
            g = grp.sort_values("N")
            axL.plot(g["N"], g["gap"], marker="o", linestyle="-", color=cmap[ratio], label=f"rd={ratio:g}")
        axL.set_title(f"Gap (train - valid) vs N  |  step_dyn={step}")
        axL.set_xlabel("N"); axL.set_ylabel("gap")
        axL.grid(True, ls="--"); axL.legend()
        axL.set_ylim(*gap_ylim)

        # 右：train / valid
        for ratio, grp in df_sel.groupby("ratio_dyn"):
            g = grp.sort_values("N")
            axR.plot(g["N"], g["final_train_loss"], marker="o", linestyle="-",  color=cmap[ratio], label=f"train rd={ratio:g}")
            axR.plot(g["N"], g["final_valid_loss"], marker="o", linestyle="--", color=cmap[ratio], label=f"valid rd={ratio:g}")
        axR.set_title(f"Train / Valid loss vs N  |  step_dyn={step}")
        axR.set_xlabel("N"); axR.set_ylabel("loss")
        axR.grid(True, ls="--"); axR.legend(ncol=2, fontsize=8)
        axR.set_ylim(*loss_ylim)

        plt.tight_layout()
        plt.show()

# =========================
# Step 6: 示例（按需启用）
# =========================
# 日志目录与枚举范围（与你当前生成命令一致）
log_dir   = "."
N_list    = [1024, 2048]                 # 如有更多，补上
step_list = [1, 2, 4]
ratio_list = ["0.0","0.1","0.2","0.4","0.8","1.0"]

candidates = make_log_filenames(N_list, ratio_list, step_list)
print("候选文件样例（前10个）:", candidates[:10])

filepaths = list_existing_logs(log_dir, candidates)
print("找到日志数量:", len(filepaths))

runs = parse_multiple_logs(filepaths)
df_final = build_final_table(runs)

# 查看汇总表（可选）
display(df_final.sort_values(["step_dyn","ratio_dyn","N"]))

# 逐张绘制，纵轴一致
plot_all_steps_static(df_final)
```
